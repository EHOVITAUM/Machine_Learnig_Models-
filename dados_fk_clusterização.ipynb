```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Carregar seus dados tratados
#df = pd.read_csv('dados_fakes_clusterizacao.csv', sep=';')

# Selecionar as colunas relevantes para clustering (por exemplo, idade e renda)
X = df[['idade', 'renda']].values

# Padronizar os dados (média zero e variância unitária)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Usar o método do cotovelo para encontrar o número ideal de clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotar o gráfico do método do cotovelo
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('Soma dos Quadrados das Distâncias (WCSS)')
plt.grid()
plt.show()

# Com base no gráfico, escolher o número ideal de clusters (no "cotovelo")
num_clusters = 6  # Substitua pelo número escolhido a partir do gráfico

# Treinar o modelo K-Means com o número ideal de clusters
kmeans = KMeans(n_clusters=num_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(X_scaled)

# Adicionar os rótulos de cluster ao DataFrame original
df['cluster'] = clusters

# Exibir informações sobre os clusters
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
for i in range(num_clusters):
    print(f'Cluster {i}:')
    print(f'Centroide (idade, renda): {cluster_centers[i]}')
    print(f'Número de pontos no cluster: {len(df[df["cluster"] == i])}\n')

# Salvar o DataFrame com os rótulos de cluster
df.to_csv('dados_fakes_clusterizados.csv', index=False, sep=';')
```
