# Feature Engineering

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
data["breast_OE"] = ordinal_encoder.fit_transform(data.breast.values.reshape(-1,1))

fig, _ = plt.subplots(1, figsize=(7,4))
ax = sns.countplot(x="breast_OE", data=data)
fig.tight_layout()
```

```python
data[["breast","breast_OE"]]
```

```python
fig, _ = plt.subplots(1, figsize=(7,4))
ax = sns.countplot(x="menopause", data=data)
fig.tight_layout()
```

```python
from sklearn.preprocessing import LabelEncoder

menopause_encoder = OrdinalEncoder()
data["menopause_OE"] = menopause_encoder.fit_transform(
    data.menopause.values.reshape(-1,1))

class_encoder = LabelEncoder()
data["Class_LE"] = class_encoder.fit_transform(
    data.Class.values.reshape(-1,1))
```

```python
irradiat_binary_encoder = OneHotEncoder(drop='first', sparse=False)
data['irradiat_BE'] = irradiat_binary_encoder.fit_transform(
    data.irradiat.values.reshape(-1,1)
)
data
```

```python
data
```

```python
dummies_age = pd.get_dummies(data['age'], prefix='ageDUM')
dummies_invnodes = pd.get_dummies(data['inv_nodes'], prefix='inv_nodesDUM')
dummies_tumorsize  =pd.get_dummies(data['tumor_size'],prefix='tumor_sizeDUM')
data = pd.concat([data, dummies_age], axis=1)
data = pd.concat([data, dummies_invnodes], axis=1)
data = pd.concat([data, dummies_tumorsize], axis=1)

data
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Supondo que você tenha um DataFrame chamado df
# Substitua 'target' pelo nome da coluna que contém os rótulos do seu conjunto de dados

# 1. Separar os rótulos do restante do DataFrame
X = data.drop('Class_LE', axis=1)
y = data['Class_LE']

# 2. Normalizar os dados
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# 3. Padronizar os dados
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# 4. Dividir os dados em conjuntos de treino e teste
X_train_norm, X_test_norm, y_train_norm, y_test_norm = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
X_train_std, X_test_std, y_train_std, y_test_std = train_test_split(X_standardized, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Modelo com dados normalizados
model_norm = LogisticRegression()
model_norm.fit(X_train_norm, y_train_norm)
predictions_norm = model_norm.predict(X_test_norm)
accuracy_norm = accuracy_score(y_test_norm, predictions_norm)
print(f'Acurácia com dados normalizados: {accuracy_norm}')

# Modelo com dados padronizados
model_std = LogisticRegression()
model_std.fit(X_train_std, y_train_std)
predictions_std = model_std.predict(X_test_std)
accuracy_std = accuracy_score(y_test_std, predictions_std)
print(f'Acurácia com dados padronizados: {accuracy_std}')
```
